```python
import matplotlib.pyplot as plt

# 你的数据
data = {
    'pt': [0.8, 0.4,0.7,0.7],#,0.8, 0.4,0.7,0.7
    'onnx': [0.8,0.5,0.8,0.8],#0.8,0.5,0.8,0.8
}

# 文本标签
categories = ['vlc', 'vlc,yolov5mweights', 'vlc+net', 'vlc+net,yolov5mweights']#,'vlc', 'vlc,5mweights', 'vlc+net', 'vlc+net,5mweights'

# 绘制柱状图
plt.figure(figsize=(10, 6))  # 设置图形大小
bar_width = 0.35  # 柱子的宽度
index = range(len(categories))  # 索引数组

# 绘制pt和onnx的柱状图
plt.bar([i + bar_width for i in index], data['pt'], bar_width, label='PT')
plt.bar([i for i in index], data['onnx'], bar_width, label='ONNX')

# 添加一些文本标签
#plt.xlabel('Category', fontsize=15)
plt.ylabel('TIME', fontsize=15)
#plt.title('Scores Comparison between PT and ONNX', fontsize=20)
plt.xticks(index, categories)  # 使用文本标签

# 添加图例
plt.legend()

# 显示图形
plt.tight_layout()  # 调整布局以适应图形
plt.show()
```

```python
import matplotlib.pyplot as plt

# 你的数据
data = {
    'pt': [0.8, 0.4,0.7,0.7],
    'onnx': [0.8,0.5,0.8,0.8]
}

# 文本标签
categories = ['a', 'b', 'c', 'd']

# 绘制柱状图
plt.figure(figsize=(10, 6))  # 设置图形大小
bar_width = 0.35  # 柱子的宽度
index = range(len(categories))  # 索引数组

# 绘制pt和onnx的柱状图
plt.bar([i for i in index], data['pt'], bar_width, label='PT')
plt.bar([i + bar_width for i in index], data['onnx'], bar_width, label='ONNX')

# 设置y轴的显示范围
#plt.ylim(10, 20)

#plt.xlabel('Category', fontsize=15)
plt.ylabel('TIME', fontsize=15)
plt.title('Scores Comparison between PT and ONNX', fontsize=20)
plt.xticks(index, categories)  # 使用文本标签
# 不显示x轴的数字标签
plt.xticks([])

# 添加图例
plt.legend()

# 显示图形
plt.tight_layout()  # 调整布局以适应图形
plt.show()
```
